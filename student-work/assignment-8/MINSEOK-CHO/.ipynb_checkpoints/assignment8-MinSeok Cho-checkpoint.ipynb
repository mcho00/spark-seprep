```python
#want to know which is the discrete and which is continuous

numeric_cols = df.select_dtypes(include=["int64", "float64"])

# Create a summary of whether the column is likely discrete or continuous
def classify_column(col):
    unique_vals = col.nunique()
    if pd.api.types.is_integer_dtype(col):
        return "Discrete" if unique_vals < 20 else "Likely Discrete"
    elif pd.api.types.is_float_dtype(col):
        return "Continuous" if unique_vals > 20 else "Possibly Discrete"
    else:
        return "Unknown"

column_classification = pd.DataFrame({
    "Type": numeric_cols.dtypes,
    "Unique Values": numeric_cols.nunique(),
    "Classification": numeric_cols.apply(classify_column)
})

column_classification

#Differences between Bounce Rate and Exit Rate: https://www.hotjar.com/google-analytics/glossary/exits/
```

# Hypothesis

## Hypothesis Establishment

Hypothesis: Returning visitors are more likely to make purchase than new visitors.
Use Revenue and VisitorType.

H0: Visitor type and purchasing behavior are independent.

H1: Visitor type and purchasing behavior are not independent (or it's correlated).


```python
visitor_cols = [col for col in df.columns if col.startswith('VisitorType_')]

visitor_purchase_counts = {vt: df[df[vt] == 1]['Revenue'].sum() for vt in visitor_cols}
visitor_purchase_series = pd.Series(visitor_purchase_counts)

visitor_purchase_series.plot(kind='bar', title='Total Purchases by Visitor Type')
plt.ylabel('Count of Revenue')
plt.xlabel('Visitor Type')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

Here, we decide to use statistical analysis through using three methods:
 * Chi-square Test
    - Chi-Square is appropriate to test two categorical variables' independence.
 * T-Test
    - T-Test will be used here to test whether returning visitors purchase more by comparing proportions.
 * One-Way Anova
    - One-Way Anova is used to test whether means of a numeric variable differ across three or more groups.

```python
# contingency table for returning visitors and revenue
contingency = pd.crosstab(df['VisitorType_Returning_Visitor'], df['Revenue'])

# Chi^2 test
chi_2, p, dof, expected = stats.chi2_contingency(contingency)

print("Chi-Squared Test: VisitorType_Returning_Visitor vs Revenue")
print(f"Chi^2 = {chi_2:.4f}, p = {p:.4f}")
print("Conclusion:", "Reject Hâ‚€ (Null)" 
      if p < 0.05 else "Fail to Reject Hâ‚€")
```

```python
# Filter only New and Returning Visitors
filtered_df = df[
    (df['VisitorType_Returning_Visitor'] == 1) |
    (df['VisitorType_New_Visitor'] == 1)
]

# Split: return and new visitors
group_returning = filtered_df[filtered_df['VisitorType_Returning_Visitor'] == 1]['Revenue']
group_new = filtered_df[filtered_df['VisitorType_New_Visitor'] == 1]['Revenue']

# t-test
t_stat, p_t = ttest_ind(group_returning, group_new, equal_var=False)

print("T-Test of Visitor Types: Returning and New Visitors")
print(f"T = {t_stat:.3f}, p-value = {p_t:.4f}")
print("Conclusion:", "Reject Hâ‚€ (Null)" 
      if p_t < 0.05 else "Fail to Reject Hâ‚€")
```

```python
# group by all visitor types
group_new = df[df['VisitorType_New_Visitor'] == 1]['Revenue']
group_returning = df[df['VisitorType_Returning_Visitor'] == 1]['Revenue']
group_other = df[df['VisitorType_Other'] == 1]['Revenue']

# One-way ANOVA
f_stat, p_val_anova = f_oneway(group_new, group_returning, group_other)

print("One-Way ANOVA: Revenue by Visitor Type")
print(f"F = {f_stat:.3f}, p-value = {p_val_anova:.4f}")
print("Conclusion:", "Reject Hâ‚€ (Null)"
      if p_val_anova < 0.05 else "â†’ Fail to Reject Hâ‚€")
```

## Hypothesis Conclusion:

Hypothesis 1 Conclusion:
Based on the results of three statistical analysis, where we rejected null values for all analysis, we will reject the null hypothesis.

There's a evidence that returning visitors are significantly more likely to make a purchase, or make revenue, compared to new visitors or other types of visitors.
